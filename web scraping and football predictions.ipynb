```python
import requests
```

```python
standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
```

```python
data = requests.get(standings_url)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data.text)
```

```python
# Using the soup object, we are selecting the css element table which is named stats_table (this was found by inspecting the
# page). We also only want to select the first table of the page.
standings_table = soup.select('table.stats_table')[0]
```

```python
# find_all allows us to find html tags while select allows to choose any html/css element present on the webpage
links = standings_table.find_all('a')
```

```python
# We want to extract on the href properties from this 'a' tag because it contains the link to each team's matches list
links = [l.get("href") for l in links]
```

```python
# filtering the links to only contain the links which point to the teams and not other players 
links = [l for l in links if '/squads/' in l]
```

```python
team_urls = [f"https://fbref.com{l}" for l in links]
```

```python
arsenal_url = team_urls[0]
arsenal_data = requests.get(arsenal_url)
```

```python
import pandas as pd

# We aim to translate the table we saw on the website into a pandas dataframe
# read_html essential looks for a string "Scores & Fixtures" from the webpage and reads the html element of the table containing
# this string
arsenal_matches = pd.read_html(arsenal_data.text, match="Scores & Fixtures")
```
